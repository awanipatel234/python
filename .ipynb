```python
fruits = {"Apple", "Banana", "Mango"}
fruits.add("Orange")
print(fruits)
```

```python
fruits = {"Apple", "Banana", "Mango"}

fruits.remove("Banana")
print(fruits)

# Uncomment below line to see error
# fruits.remove("Grapes")   # KeyError
```

```python
fruits = {"Apple", "Banana", "Mango"}

fruits.discard("Mango")
print(fruits)

fruits.discard("Grapes")   # No error
print(fruits)
```

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

union_set = set1.union(set2)
print("Union of sets:", union_set)
```

```python
A = {1, 2, 3}
B = A.copy()

B.add(4)

print("Original set:", A)
print("Copied set:", B)
```
