```python
def check_even_odd(number):
    if number % 2 == 0:
        print("Even number")
    else:
        print("Odd number")
check_even_odd(7)
```

```python
def maximum(a, b):
    if a > b:
        return a
    else:
        return b 
    result = maximum(10, 20)
    print("Maximum:", result)
```

```python
def average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count

nums = [10, 20, 30, 40, 50]
avg = average(nums)
print("Average:", avg)
```

```python
def count_vowels(string):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in string:
        if char in vowels:
            count += 1
    return count
sentence = "hello_world!"
vowel_count = count_vowels(sentence)
print("Number of vowels:", vowel_count)
```

```python
def factorial(n):
    if n < 0:
        return "factorial does not exist  for negative numbers"
    result = 1
    for i in range(1, n + 1):
        result *= i
        
    return result
num = 5
print("Factorial of", num, "is", factorial(num))
    
```

```python
def is_palindrome(s):
    s = s.replace(" ", "").lower()
    length = len(s)
    for i in range(length // 2):
        if s[i] != s[length - i - 1]:
            return False
    return True
word = "racecar"
print(f"Is"'{word}'" a palindrome? {is_palindrome(word)}")
```

```python
def simple_interest(principal, time,rate=5):
    """Calculate simple interest.
    SI = (P * T * R) / 100
    Default rate is 5%.
    
    
    """
    si = (principal * time * rate) / 100
    return si
p = 1000
t = 2
print(f"Simple interest for principle{p} over{t} years is {simple_interest(p, t)}")
print(f"with custom rate 8%: {simple_interest(p, t, rate=8)}")
    
```
